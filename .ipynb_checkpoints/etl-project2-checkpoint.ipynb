```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Data Extracted from Contract Research Organization of a Pharmaceuticle 
patient_file = "Resources/Patient Data_1.csv"
patient_df = pd.read_csv(patient_file)
patient_df.head()
```

```python
#Data Extracted from Contract Research Organization of a Pharmaceuticle 
patient_key = "Resources/Patient Data_key.csv"
patientkey_df = pd.read_csv(patient_key)
patientkey_df.head(25)
```

```python
# Transform patient DataFrame
# Create a filtered dataframe from specific columns
patientdata_cols = ["patientid", "Treatment", "visitc", "batch", "qc", "gate_path", "cell_prop", "cell_count"]
patientdata_transformed= patient_df[patientdata_cols].copy()

# Rename the column headers
patientdata_NewColumnHeader = patientdata_transformed.rename(columns={"patientid": "patientid",
                                                          "Treatment": "treatment",
                                                          "visitc":"sampleday",
                                                          "batch":"analysisbatch",
                                                          "qc": "qualitycontrol",
                                                          "gate_path":"celltype",
                                                          "cell_prop":"perecentcells",
                                                          "cell_count":"cellcount"})


patientdata_NewColumnHeader.head()
```

```python
# Clean the data by dropping Analysis Records that failed Quality Control 
patientdata_NewColumnHeader[patientdata_NewColumnHeader.qualitycontrol != 'FAIL']
patientdata_NewColumnHeader.head()

patientdata_newIndex = patientdata_NewColumnHeader.set_index('patientid')
patientdata_newIndex.head()
```

```python
# Transform patientkey DataFrame

# Rename the column headers
patientkeydata_NewColumnHeader = patientkey_df.rename(columns={"patientid": "patientid",
                                                          "sex": "patientsex",
                                                          "age":"patientage",
                                                          "CORTSTER": "onsteroids",
                                                          "ethniciy": "patientethnicity"})


patientkeydata_NewColumnHeader.head()
```

```python
#set index to patient ID
patientkeydata_newIndex = patientkeydata_NewColumnHeader.set_index('patientid')
patientkeydata_newIndex.head()
```

```python
#Create database connection
connection_string = "postgres:postgres@localhost:5432/etlproject2_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm the two tables created in SQL
engine.table_names()
```

```python
#Load DataFrames into database
patientdata_newIndex.to_sql(name='patient_data', con=engine, if_exists='append', index=True)
patientkeydata_newIndex.to_sql(name='patient_key', con=engine, if_exists='append', index=True)
```
