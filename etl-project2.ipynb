```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Data Extracted from Contract Research Organization of a Pharmaceuticle 
patient_file = "Resources/Patient Data_1.csv"
patient_df = pd.read_csv(patient_file)
patient_df.head()
```

```python
#Data Extracted from Contract Research Organization of a Pharmaceuticle 
patient_key = "Resources/Patient Data_key.csv"
patientkey_df = pd.read_csv(patient_key)
patientkey_df.head(25)
```

```python
# Transform patient DataFrame
# Create a filtered dataframe from specific columns
patientdata_cols = ["patientid", "Treatment", "visitc", "batch", "qc", "gate_path", "cell_prop", "cell_count"]
patientdata_transformed= patient_df[patientdata_cols].copy()

# Rename the column headers
patientdata_NewColumnHeader = patientdata_transformed.rename(columns={"patientid": "Patient ID",
                                                          "Treatment": "Treatment",
                                                          "visitc":"Sample Day",
                                                          "batch":"Analysis Batch",
                                                          "qc": "QualityControl",
                                                          "gate_path":"Cell Type",
                                                          "cell_prop":"% Of Total Cells",
                                                          "cell_count":"Cell Count"})


patientdata_NewColumnHeader.head()
```

```python
# Clean the data by dropping Analysis Records that failed Quality Control 
patientdata_NewColumnHeader[patientdata_NewColumnHeader.QualityControl != 'FAIL']
patientdata_NewColumnHeader.head()
```

```python
# Transform patientkey DataFrame

# Rename the column headers
patientkeydata_NewColumnHeader = patientkey_df.rename(columns={"patientid": "Patient ID",
                                                          "sex": "Patient Sex",
                                                          "age":"Patient Age",
                                                          "CORTSTER":"On Steroids?",
                                                          "ethniciy": "Patient Ethnicity"})


patientkeydata_NewColumnHeader.head()
```

```python
#Create database connection
connection_string = "postgres:postgres@localhost:5432/etlproject2_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm the two tables created in SQL
engine.table_names()
```

```python
#Load DataFrames into database
patientdata_NewColumnHeader.to_sql(name='patient_data', con=engine, if_exists='append', index=True)
patientkeydata_NewColumnHeader.to_sql(name='patient_key', con=engine, if_exists='append', index=True)
```
